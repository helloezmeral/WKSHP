## Model Deployment

After the model development and testing, we are now ready to deploy the model, and start serving.

### Model Registry

First step towards model deployment is to register the model. 
In EPIC, navigate to the Model Registry. In the Model Management screen, click on <font color='green'>Register New Model</font>  and register the model as shown below. Make sure to give the version number as your student number (e.g. 11) to enable registering and deploying a model that's unique to your work. Select <font color='green'>Browse</font> to choose the model directory path for the pickle file and the code directory for the scoring script. 

![ModelRegistry](Pictures/ModelRegistry.png)

### Model Deployment

Deploy the newly registered model by clicking the Deploy icon (jewel), under Actions, for the Income_Prediction_## model. This method automatically pre-selects the Income_Prediction_## model and its version from model registry to deploy.

![Model Deployment](Pictures/ModelDeployment.png)

Ensure that the Deployment cluster status appears as <font color='green'>Ready</font>. It may 1-2 minutes for the Deployment cluster to be created.

**NOTE:** Alternately, the other method to create Deployment cluster is to navigate to the Deployments screen and click on <font color='green'>Create Deployment</font> and enter a name for Deployment cluster and choose the model name and version from the dropdown. You may leave the rest of the configuration settings as default.

### Model Serving

To generate income prediction requests, we will use the **curl** command to post a prediction request. Open a terminal window in your Jupyter notebook by clicking + on your Notebook controls to invoke the Launcher and click on Terminal.

![Notebook Terminal](Pictures/NotebookTerminal.png)

The below curl command should be copied to your terminal window and make the two changes in the steps below. You can copy the curl command in your favorite text editor for editing and create a bash script to test, or just edit this markdown by double-clicking the cell and edit it inline, then paste the entire curl command in the terminal window.

Navigate to the Deployments screen and click on the <font color='green'>Income_Prediction_##</font> link. You will see a list all the nodes (containers) running for this deployment cluster with various roles. 

![Model Serving](Pictures/ModelServing.png)
Copy the Model Serving Load Balancer URL and paste it in the **curl** command shown below for the post request for model serving.

![AuthToken](Pictures/AuthToken.png)
Hover over and Click on the <font color='green'>Auth Token</font>, it gets copied to the clipboard. Paste it in the **curl** command in the header Auth-Token shown below.

curl --location -v -k --request POST 'https://gateway2.etc.fr.comm.hpecorp.net:10523/Income_Prediction_1/1/predict' \
--header 'X-Auth-Token: xTsgXvX3PR' \
--header 'Content-Type: application/json' \
--data-raw '{
    "use_scoring": true,
    "scoring_args": {
        "age": 35,
        "workclass": "Self-emp-inc",
        "fnlwgt": 182148,
        "education": "Bachelors",
        "education_num": 13,
        "marital_status": "Married-civ-spouse",
        "occupation": "Exec-managerial",
        "relationship": "Husband",

}'| python -m json.tool
